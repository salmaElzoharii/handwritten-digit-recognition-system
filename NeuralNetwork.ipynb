```python
from torchvision import datasets,transforms
from sklearn.model_selection import train_test_split
test_data= datasets.MNIST(
    root='/archive',
    train= True,
    download=True,
    transform=transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))])
)
train_data= datasets.MNIST(
    root='/archive',
    train= True,
    download=False,
    transform=transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))])
)
train_set, val_set = train_test_split(train_data, test_size=0.2, random_state=42, stratify=train_data.targets)


from torch.utils.data import DataLoader
# Create DataLoader for train data and valdation data
train_loader = DataLoader(train_set, batch_size=64, shuffle=True)
val_loader = DataLoader(val_set, batch_size=64, shuffle=False)
# Create DataLoader for test data
test_loader = DataLoader(test_data, batch_size=64, shuffle=False)
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split
from torchvision import datasets, transforms
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt


# Different batch sizes and learning rates to try
batch_sizes = [32, 64, 128, 256, 512]
learning_rates = [0.1, 0.01, 0.001, 0.0001, 0.00001]
num_epochs = 10

# Loop over learning rates
for lr in learning_rates:
    print(f'\nLearning Rate: {lr}')

    # Loop over batch sizes
    for batch_size in batch_sizes:
        print(f'\tBatch Size: {batch_size}')

        # DataLoader for efficient loading with the current batch size
        train_loader = DataLoader(train_set, batch_size=batch_size, shuffle=True)
        val_loader = DataLoader(val_set, batch_size=batch_size, shuffle=False)

        # Neural Network architecture with dropout and layer normalization
        class NeuralNetwork(nn.Module):
          def forward(self, x):
            x = x.view(-1, 28 * 28)
            x = self.relu(self.fc1(x))

            x = self.dropout1(x)  # Apply dropout
            x = self.relu(self.fc2(x))

            x = self.relu(self.fc3(x))
            x = self.fc4(x)
            return x

          def __init__(self):
            super(NeuralNetwork, self).__init__()
            self.fc1 = nn.Linear(28 * 28, 256)

            self.relu = nn.ReLU()
            self.dropout1 = nn.Dropout(0.2)  # Dropout with a dropout probability of 0.2
            self.fc2 = nn.Linear(256, 128)

            self.fc3 = nn.Linear(128, 64)
            self.fc4 = nn.Linear(64, 10)




        # Initialize the model with dropout and layer normalization
        model = NeuralNetwork()

        # Train the model
        optimizer = optim.SGD(model.parameters(), lr=lr)
        criterion = nn.CrossEntropyLoss()
        train_losses, val_losses, train_acc, val_acc = [], [], [], []

        for epoch in range(num_epochs):
            model.train()
            total_train_loss = 0.0
            correct_train = 0

            for images, labels in train_loader:
                optimizer.zero_grad()
                outputs = model(images.view(images.shape[0], -1))
                loss = criterion(outputs, labels)
                loss.backward()
                optimizer.step()

                total_train_loss += loss.item()
                _, predicted = torch.max(outputs.data, 1)
                correct_train += (predicted == labels).sum().item()

            avg_train_loss = total_train_loss / len(train_loader.dataset)
            train_accuracy = correct_train / len(train_loader.dataset)

            train_losses.append(avg_train_loss)
            train_acc.append(train_accuracy)

            # Validation
            model.eval()
            total_val_loss = 0.0
            correct_val = 0

            with torch.no_grad():
                for images, labels in val_loader:
                    outputs = model(images.view(images.shape[0], -1))
                    loss = criterion(outputs, labels)
                    total_val_loss += loss.item()
                    _, predicted = torch.max(outputs.data, 1)
                    correct_val += (predicted == labels).sum().item()

            avg_val_loss = total_val_loss / len(val_loader.dataset)
            val_accuracy = correct_val / len(val_loader.dataset)

            val_losses.append(avg_val_loss)
            val_acc.append(val_accuracy)

            print(f'Epoch [{epoch + 1}/{num_epochs}], '
                  f'Training Loss: {avg_train_loss:.4f}, Training Accuracy: {train_accuracy:.4f}, '
                  f'Validation Loss: {avg_val_loss:.4f}, Validation Accuracy: {val_accuracy:.4f}')

        # Plotting
        plt.figure(figsize=(10, 5))
        plt.subplot(1, 2, 1)
        plt.plot(train_losses, label='Training Loss')
        plt.plot(val_losses, label='Validation Loss')
        plt.legend()
        plt.title(f'Training and Validation Loss (LR: {lr}, Batch Size: {batch_size})')

        plt.subplot(1, 2, 2)
        plt.plot(train_acc, label='Training Accuracy')
        plt.plot(val_acc, label='Validation Accuracy')
        plt.legend()
        plt.title(f'Training and Validation Accuracy (LR: {lr}, Batch Size: {batch_size})')

        plt.show()

        # Report final accuracy
        print(f'\tFinal Training Accuracy: {train_acc[-1]:.4f}')
        print(f'\tFinal Validation Accuracy: {val_acc[-1]:.4f}')

        # Model evaluation
        model.eval()
        with torch.no_grad():
            for images, labels in val_loader:
                outputs = model(images.view(images.shape[0], -1))
                _, predicted = torch.max(outputs, 1)

            # Display the images and predictions
            plt.figure(figsize=(15, 5))
            for i in range(min(5, len(images))):
                plt.subplot(1, 5, i + 1)
                plt.imshow(images[i][0], cmap='gray')
                plt.title(f'Predicted: {predicted[i]}, Actual: {labels[i]}')
                plt.axis('off')
            plt.show()
```
